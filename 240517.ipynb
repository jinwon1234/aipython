```python
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]
fig, axs = plt.subplots(2,2)
axs[0,0].plot(x,y)
axs[0,1].scatter(x,y)
axs[1,0].bar(x,y)
axs[1,1].barh(x,y)
fig.suptitle('Drawing Graph')
plt.show()
```

```python
import seaborn as sns
sns.get_dataset_names()
```

```python
iris_data=pd.read_csv('iris.csv')
sns.scatterplot(x='petal_length',y='petal_width'
,hue='species',data= iris_data)
plt.title('Scatter Plot by seaborn', fontsize=20)
plt.show()
```

```python
data = pd.DataFrame({
'x': [1, 2, 3, 4, 5],
'y': [10, 20, 15, 25, 30],
'category': ['A', 'B', 'A', 'B', 'A']
})
sns.scatterplot(x='x',y='y',hue='category', data = data)
```

```python
data = pd.DataFrame({
'category': ['A', 'B', 'A', 'C', 'B', 'A', 'A', 'C', 'B', 'B'],
'value': [10, 20, 15, 25, 30, 12, 18, 22, 28, 35]
})
sns.countplot(x='category', hue='category',data=data)
```
