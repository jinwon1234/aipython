```python
minute = int(input())
if minute < 60:
  print('{:d}분'.format(minute))
else:
  print('{:d}시간 {:d}분'.format(minute // 60, minute % 60))
```

```python
c_t = int(input('섭씨 온도를 입력하시오'))
f_t = c_t * 1.8 + 32
print('섭씨 %d 도는 화씨 %.1f도 입니다.' % (c_t, f_t))
```

```python
r = int(input('원의 지름을 입력하시오'))
PI = 3.14
circumference = r*2*PI
area = PI * (r ** 2)
print('반지름이 {} 인 원의 둘레는 {}이고 면적은 {} 입니다.'.format(r,round(circumference,1),round(area,1)))
```
